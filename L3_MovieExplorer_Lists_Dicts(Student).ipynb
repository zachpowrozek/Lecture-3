
## üìñ Introduction
Today we‚Äôll learn how to organize data with **lists** and **dictionaries**.  

- **Lists** store items in order: `["Inception", "Titanic", "Avatar"]`  
- **Dictionaries** store labeled data: `{"title": "Inception", "rating": 8.8}`  
- **APIs** often give us **lists of dictionaries** (JSON).  


## 1Ô∏è‚É£ Lists (Hardcoded Example)


Lists use **square brackets `[ ]`**.  
Each item in the list is called an **element**.  
Elements have a position number called an **index**.  


````python
movies = ["Inception", "Titanic", "Avatar"]
print(movies)
````


üìä **Visual of a list with indices**:

```
List:   ["Inception",    "Titanic",    "Avatar"]
Index:       0              1             2
```

üëâ Lists **start at 0**.  
üëâ The last index is **length ‚àí 1**.  


````python
print(f"First movie: {movies[0]}")
print(f"Second movie: {movies[1]}")
print(f"Third movie: {movies[2]}")

print(f"There are {len(movies)} items in the list.")
````

## 2Ô∏è‚É£ Dictionaries (Hardcoded Example)


Dictionaries use **curly braces `{ }`** and store data in **key ‚Üí value pairs**.  


````python
import pprint

movie1 = {
    "title": "Inception",
    "year": 2010,
    "rating": 8.756,
    "votes": 1234567,
    "video": True
}

pprint.pprint(movie1)


# Uncomment the lines of code below.
# Don't forget, you can select all of the lines and use Cmd or Ctr plus "/"

print("\n")     # Add a blank line
print(f"Title: {movie1['title']}")
print(f"Year: {movie1['year']:,}")
print(f"Rating: {movie1['rating']:.1f}")
print(f"Votes: {movie1['votes']:,}")
````

üìä Visual of a dictionary:
```python
"title"  ‚Üí  "Inception"
"year"   ‚Üí  2010
"rating" ‚Üí  8.756
"votes"  ‚Üí  1234567
"video"  ‚Üí  True
```

üëâ Dictionaries use keys (like "title") instead of positions.

## ü§î When to Use Lists vs. Dictionaries


- **List** ‚Üí use when you think about the data **linearly**.  
- **Dictionary** ‚Üí use when you want **labeled info** about one thing.  
- **List of dictionaries** ‚Üí use for **many items with details**.

## A little thing I wrote for you to make long strings fit on the page.

Go ahead and run the cell below.  You don't have to know how it works.

````python
# üì¶ Setup cell to make print() wrap text in Colab
import builtins, textwrap

def wrapped_print(*args, width=80, **kwargs):
    """
    Works just like the normal print(), but long text
    will automatically wrap to the given width.
    """
    new_args = [textwrap.fill(str(a), width=width) for a in args]
    builtins.print(*new_args, **kwargs)

# Replace the default print with our wrapped version
print = wrapped_print

print("‚úÖ Print has been updated! Long lines will now wrap automatically.")
````
